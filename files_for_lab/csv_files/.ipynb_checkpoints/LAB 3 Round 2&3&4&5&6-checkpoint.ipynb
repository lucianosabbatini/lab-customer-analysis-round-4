```python
#Assigning Nulls to median
df['Number of Open Complaints']=df['Number of Open Complaints'].fillna(df['Number of Open Complaints'].median())
```

```python
df['Vehicle Class'].value_counts(dropna=False)
```

```python
#Assigning Null to most common 4 door car
df['Vehicle Class'].fillna('Four-Door Car', inplace=True)
```

```python
df['Vehicle Size'].value_counts(dropna=False)
```

```python
#Assigning nulls to most common medsize
df['Vehicle Size'].fillna('Medsize', inplace=True)
```

```python
X_pred
```

#### Train-Test Split

```python
#<train-Testing
y=numerical['Total Claim Amount']
X=X_pred
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#X_train.head()
```

```python
from sklearn import linear_model
```

#### Apply linear regression.

```python
m = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
# Checking R2 score for the model on training
from sklearn.metrics import r2_score  
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

# Model Validation

#### R2 Score

```python
#Calculating the predicted R2 Score on the test data
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
predictions_test[:5]
```

#### MSE

```python
from sklearn.metrics import mean_squared_error
#Calculating MSE
mse=mean_squared_error(y_test,predictions_test)
mse
```

#### RMSE

```python
#Calculating RMSE
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))  
rmse
```
